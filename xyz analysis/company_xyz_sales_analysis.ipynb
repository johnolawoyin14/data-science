```python
df=pd.read_csv("branches_folder/Abuja_Branch.csv")
df
```

```python
files=[files for files in listdir("branches_folder")]
all_branches=pd.DataFrame()
for file in files:
    df=pd.read_csv(f"branches_folder/{file}")
    all_branches=pd.concat([all_branches,df])
all_branches.to_csv("branches_folder/all_branches.csv",index=False)
```
